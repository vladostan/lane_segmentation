```python
hex_colors = ['#000000','#4A4A4A','#FF0000', '#407700', '#00FF37', '#8B572A', '#0000FF', '#FF7600', '#50E3C2']
```

```python
rgb_colors = []

for hex_color in hex_colors:
    rgb_colors.append(webcolors.hex_to_rgb(hex_color))
```

```python
def colorize(mask, colors):
    colorMask = np.zeros([mask.shape[0], mask.shape[1], 3],dtype=np.uint8)
    for r in range(mask.shape[0]):
        for c in range(mask.shape[1]):
            colorMask[r,c,] = colors[mask[r,c]]

    return colorMask
```

```python
r = np.random.randint(80, size=3)

fig, axes = plt.subplots(nrows=2, ncols=3)
fig.set_size_inches(20,10)

axes[0,0].imshow(x[r[0]])
axes[0,1].imshow(x[r[1]])
axes[0,2].imshow(x[r[2]])

axes[1,0].imshow(colorize(y[r[0]], colors = np.array(rgb_colors)))
axes[1,1].imshow(colorize(y[r[1]], colors = np.array(rgb_colors)))
axes[1,2].imshow(colorize(y[r[2]], colors = np.array(rgb_colors)))

fig.tight_layout()
```
